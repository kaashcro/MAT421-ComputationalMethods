**Interpolation Problem Statement**  



Interpolation is a way to look at data that we dont nessassaraly have to data for and make graphs make sense from them

```python
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
```

```python
x=np.linspace(0,10,11) #setting up data for interpolation
y=[0,1,3,6,3,0,1,5,7,10,15]
plt.plot(x,y, 'ko')
plt.xlabel('Time(s)')
plt.ylabel('Velocity (m/s)')
plt.title("acceleration of a car")
```

Linear Interpolation


```python
f = interp1d(x, y) #finding a specific data point in the data set
y_velocity=f(3)
print(y_velocity)
```

```python
plt.figure(figsize = (11,15)) #Marking the data point in the data set
plt.plot(x, y, '-ob')
plt.plot(3, y_velocity, 'ro')
plt.title('Linear Interpolation at x = 3')
plt.xlabel('Time(s)')
plt.ylabel('Velocity (m/s)')
plt.show()
```

Cubic Spline Interpotaltion

```python
f = CubicSpline(x, y, bc_type='natural') #setting up a curve of best fit for the data set that you don't know
x_new = np.linspace(0, 11, 100)
y_new = f(x_new)
plt.figure(figsize = (11,15))
plt.plot(x_new, y_new, 'b')
plt.plot(x, y, 'ro')
plt.title('Cubic Spline Interpolation')
plt.xlabel('Time(s)')
plt.ylabel('Velocity (m/s)')
plt.show()
```

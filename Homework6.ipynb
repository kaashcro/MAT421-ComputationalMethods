**Continuity and Differentiation**

Continuous function is a function that does not have any abrupt changes
in value, known as discontinuities.

```python
!pip install sympy
import numpy as np
import sympy
```

```python
import matplotlib.pyplot as plt
```

```python
#an example of a continuous function, where there are no sharp points of disconnect
x=np.linspace(0,5,500)
y=np.sin(2*x)+np.cos(5*x)
plt.plot(x,y)
```

```python
#finding the limit of the function
x = sympy.Symbol('x')
sympy.limit(sympy.sin(2*x)+sympy.cos(5*x), x, 0)
```

The derivative of a function of a real variable measures the sensitivity to
change of the function value (output value) with respect to another variable

```python
#The derivative of my equation
x=np.linspace(0,5,500)
h=0.01
diff_eq=np.diff(y)/h
x_diff = x[:-1:]
real_eq = 2*np.cos(2*x_diff) - 5*np.sin(5*x_diff)
plt.plot(x_diff, diff_eq, '--', label = 'computer solved equation')
plt.plot(x_diff, real_eq,label = 'Exact Solution')
plt.legend()
plt.show()
```

Taylors Theorem

Taylors theorem gives an approximation of a differentiable function around
a given point by a polynomia

```python
from sympy import *
#Finding the Talor series for my equation
x = sympy.Symbol('x')
g= x
eq=sin(2*g) + cos(5*x)
taylorExpand = eq.series(x, 0, 3).removeO()
display(taylorExpand)

p1 = plot(eq, show=False)
p2 = plot(taylorExpand, show=False)
p1.append(p2[0])
print('Taylor expansion shows ok behavior near the 0')
p1.show()
```
